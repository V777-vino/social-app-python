```python
import social_app_functions as s
import Register_validation as v
import getpass
import re


def home():
    print("ðŸŽ‡ Welcome to Social app ðŸŽ‡")
    print("\npress '1' for Sign up to create a New account")
    print("\npress '2' for Login")
    print("\npress '3' for Exit")
    num = int(input("\nEnter your choice:"))
    
    if(num==1):
        fname = input("\nEnter your first name:")
        lname = input("\nEnter your last name:")
        email = input("\nEnter your mail id:")
        phone_num = input("\nEnter your phone number:")
        gender = input("\nEnter the gender(M/F):")
        dob = input("\nEnter the date of birth(YYYY-MM-DD):")
        password = input("\nSet your password:")
        if(v.name_validation(fname) == True and s.is_registered(email)==True and v.name_validation(lname)== True and v.mail_validation(email)==True and v.phone_number_validation(phone_num) == True and v.gender_validation(gender) == True and v.password_validation(password) == True and v.is_valid_date(dob)==True):
            s.insert_user_details(fname,lname,email,phone_num,gender,dob,password)
            print("\nUser account created")
            print("==============================================")
        home()
    elif(num==2):
        mail = input("\nEnter the mail id:")
        password = input("\nEnter the password:")
        if(s.login(mail,password)==True):
            print("\nLogged in!!")
            print("=====================================================")
            menu()
        else:
            print("\nInvalid credentials...Try again!")
            print("=======================================================")
            home()
    elif(num==3):
        print("\nThanks for using Social appðŸ‘")
        print("=========================================================")
    else:
        print("\nPlease enter the valid choice")
        home()
def menu():
    print("\n=>Press '1' to see all posts")
    print("\n=>Press '2' for update mail")
    print("\n=>Press '3' for show user details by user id")
    print("\n=>Press '4' for update password")
    print("\n=>Press '5' for delete post")
    print("\n=>Press '6' for update phone number")
    print("\n=>Press '7' for delete user by user id")
    print("\n=>Press '8' to see your post by user id")
    print("\n=>Press '9' to post a message:")
    print("\n=>Press '10' for search user by their name")
    print("\n=>Press '11' fro Exit")
    choice = int(input("\nPlease Enter your choice:"))
    
    if(choice ==1):
        s.see_all_posts()
        menu()
    elif(choice==2):
        user_id = int(input("\nEnter your user id:"))
        mail = input("\nEnter your new mail id:")
        if(v.mail_validation(mail)==True):
            s.update_mail_id(user_id,mail)
        menu()
    elif(choice==3):
        user_id = int(input("\nEnter your user id:"))
        s.show_user_detail_by_id(user_id)
        menu()
    elif(choice==4):
        user_id = int(input("\nEnter your user id:"))
        password = getpass.getpass(prompt="Enter your password:")
        if(v.password_validation(password)==True):
            s.update_password(user_id,password)
        menu()
    elif(choice == 5):
        post_id = int(input("\nEnter your post id:"))
        s.delete_post(post_id)
        menu()
    elif(choice==6):
        user_id = int(input("\nEnter your user id:"))
        phone = input("\nEnter your new phone number:")
        if(v.phone_number_validation(phone)==True):
            s.update_phone_number(phone)
        menu()
    elif(choice ==7):
        user_id = int(input("\nEnter your user id:"))
        s.delete_user(user_id)
        menu()
    elif(choice==8):
        user_id = int(input("\nEnter your user id:"))
        s.your_posts(user_id)
        menu()
    elif(choice ==9):
        user_id = int(input("\nEnter your user id:"))
        post = input("\nEnter your message:")
        s.post_message(user_id,post)
        menu()
    elif(choice == 10):
        name = input("\nEnter the name:")
        s.search_user_by_name(name)
        menu()
    elif(choice == 11):
        print("\nExit")
        home()
    else:
        print("\nEnter the correct choice")
        menu()
        
        
home()
```
